```python
import random
import time

text = """What is the difference between
a ' and a "? Or between a " and a \"?\n
One is what we see when we're typing our program.
The other is what appears on the "console."
"""
print('Question 1: ')
print(text, '\n')

print('Question 2:')

print('Generating Random Sequence...')
time.sleep(3)
seq_length = 43
rand_seq = ''
for ind in range(seq_length):
    rand_nuc = random.choice("ATGC")
    rand_seq += rand_nuc
    

def get_reverse_complement(seq):
    rc_mapper = {'A': 'T', 'T': 'A', 'G': 'C', 'C': 'G'}
    rc = ''.join([rc_mapper[nuc] for nuc in seq[::-1]])
    rc = rc.replace('T', 'U')
    
    return rc

seq_rna = rand_seq.replace('T', 'U')
rc_rna = get_reverse_complement(rand_seq)
print('Sequence: ', rand_seq)
print('RNA Seq:  ', seq_rna)
print('RC RNA:   ', rc_rna, '\n')

file_name = 'rand_seq.txt'
with open(file_name, 'w') as seq_file:
    seq_file.write(rand_seq)

seq = open(file_name, 'r').read()
    
    
def get_nucleotide_frequency(seq):
    GC = 0
    AT = 0
    seq

    for nuc in seq:
        if nuc in 'AT':
            AT += 1
        elif nuc in 'GC':
            GC += 1
    # get ratio
    GC /= len(seq) 
    AT /= len(seq) 
    
    # multiply to get percentile
    GC *= 100
    AT *= 100

    return GC, AT
   
gc, at = get_nucleotide_frequency(seq)   
gc_text = f'The GC content of the DNA sequence given is: {round(gc, 1)}%'
at_text = f'The AT content of the DNA sequence given is: {round(at, 1)}%'

print('Question 3: ')
print('Calculating GC content...')
time.sleep(2)
print(gc_text)
print(at_text)

with open('DNA_Statistics.txt', 'w') as stats_file:
    for txt in [gc_text, at_text]:
        stats_file.write(txt + '\n')
```
